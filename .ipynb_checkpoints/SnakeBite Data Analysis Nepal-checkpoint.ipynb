```python
import pandas as pd
```

```python
df=pd.read_csv("Data/snake-bites-cases-and-deaths-nepal-2065-66-to-2073-74.csv") #load data
```

```python
df.head()  #display first few rows of data
```

```python
df.info()  #display structure and basic information of data
```

```python
df.describe()  #statistics for numerical columns
```

```python
import matplotlib.pyplot as plt
```

```python
#Visualizing snake bite trend over the years
plt.figure(figsize=(10, 6))
plt.plot(df['Fiscal year'], df['Total cases'], marker='o')
plt.xlabel('Fiscal year')
plt.ylabel('Total Cases')
plt.title('Total Snake Bite Cases Over the Years')
plt.grid(True)
plt.show()
```

```python
df['cure rate']=df['Cure']/df['Total cases']  #Visualizing cure rate
```

```python
df['cure rate'],df['Fiscal year']
plt.figure(figsize=(10,6))
plt.plot(df['Fiscal year'],df['cure rate'],marker='o',color='green')
plt.xlabel('Fiscal year')
plt.ylabel('cure rate')
plt.grid(True)
plt.show()
```

```python
# Calculate percentage of deaths
df['Death Percentage'] = (df['No. deaths'] / df['Total cases']) * 100

# Display average death percentage
average_death_percentage = df['Death Percentage'].mean()
print(f"Average Death Percentage: {average_death_percentage:.2f}%")

# Plot death percentage over the years
plt.figure(figsize=(10, 6))
plt.plot(df['Fiscal year'], df['Death Percentage'], marker='x', color='red')
plt.xlabel('Fiscal year')
plt.ylabel('Death Percentage (%)')
plt.title('Death Percentage Over the Years')
plt.grid(True)
plt.show()
```

```python
# Calculate average number of poisonous and non-poisonous bites per year
avg_poisonous = df['Poisonous'].mean()
avg_non_poisonous = df['Non-Poisonous'].mean()

print(f"Average Poisonous Bites per Year: {avg_poisonous:.2f}")
print(f"Average Non-Poisonous Bites per Year: {avg_non_poisonous:.2f}")
```
